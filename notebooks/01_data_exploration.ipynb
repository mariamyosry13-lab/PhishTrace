```python
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')

df = pd.read_csv('../data/processed/phishtrace_dataset.csv')
print(f"Total URLs : {len(df):,}")
print(f"Phishing   : {(df['label']==1).sum():,}  ({(df['label']==1).mean():.1%})")
print(f"Legitimate : {(df['label']==0).sum():,}  ({(df['label']==0).mean():.1%})")
df.head()
```

```python
df['url_length'] = df['url'].str.len()

fig, axes = plt.subplots(1, 2, figsize=(12, 4))
df[df['label']==1]['url_length'].hist(bins=50, ax=axes[0], color='#e63946', alpha=0.7)
axes[0].set_title('Phishing URL Lengths')
axes[0].set_xlabel('Length (chars)')

df[df['label']==0]['url_length'].hist(bins=50, ax=axes[1], color='#2ec4b6', alpha=0.7)
axes[1].set_title('Legitimate URL Lengths')
axes[1].set_xlabel('Length (chars)')

plt.tight_layout()
plt.savefig('../reports/figures/url_length_dist.png', dpi=120)
plt.show()
print(f"Avg phishing URL length   : {df[df['label']==1]['url_length'].mean():.0f} chars")
print(f"Avg legitimate URL length : {df[df['label']==0]['url_length'].mean():.0f} chars")
```
